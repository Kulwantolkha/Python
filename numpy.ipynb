####For particular rows and columns

We can use the axis parameter to get the max, min and sum of a particular column or row


*   axis=0 for the column
*   axis=1 for the row

```python
arr = np.random.randint(2,10,size= (3,5))
print(arr)
print("-"*50)
#For rows
print("Row wise: ")
print("MAX: ",arr.max(axis=1))
print("MIN: ",arr.min(axis=1))
print("SUM: ",arr.sum(axis=1))
print("-"*50)
#For columns
print("Column wise: ")
print("MAX: ",arr.max(axis=0))
print("MIN: ",arr.min(axis=0))
print("SUM: ",arr.sum(axis=0))
```

now here the seed value is defined for **np.random** so whenever we define **arr** and **arr1** we need to define in such a way that suits the seed function and it will generate that same value until and unless we change that seed value

#Sorting Funtion

```python
np.random.seed(10)
arr = np.random.randint(1,10,(3,5))
print(arr)
print("-"*50)
print(np.sort(arr))
#we can sort row and column wise as well
print("-"*50)
print("Column wise: ")
print(np.sort(arr,axis = 0))
print("-"*50)
print("Row wise: ")
print(np.sort(arr,axis = 1))
```

```python
#We can also do sorting like
arr = arr.flatten()
print(np.sort(arr))
print(np.sort(arr.flatten()))
arr = np.sort(arr.flatten())
#now can reshape this in original form
arr = arr.reshape(3,5)
print(arr)

# arr = np.reshape(np.sort(arr.flatten()),arr.shape)
```

#Mean, Variance and Standard deviation

```python
arr = np.random.randint(1,10,10)
print(arr)
print("-"*50)
print(np.mean(arr))
print("-"*50)
print(np.average(arr))
print("-"*50)
print(np.median(arr))
print("-"*50)
print(np.var(arr))
print("-"*50)
print(np.std(arr))
```

#Where Function

```python
arr = np.random.randint(1,100,10)
print(arr)
print("-"*50)
print(np.where(arr>50))                    #gives the index
#to change the  array values that are greater than 50 into 100
# arr[arr>50] = 100
print(arr)
#or
print("-"*50)
print(np.where(arr%2==0,100,arr))
print(np.where(arr%2==0,'even',arr))
```

#Merging array

```python
arr1 = np.random.randint(1,10,(2,4))
arr2 = np.random.randint(1,10,(2,4))
print(arr1)
print("-"*50)
print(arr2)
print("-"*50)
print(np.concatenate((arr1,arr2)))                      #default axis is 0
print("-"*50)
print(np.concatenate((arr1,arr2),axis=1))
print("-"*50)
print(np.vstack((arr1,arr2)))                            #vertical stack (axis = 0)
print("-"*50)
print(np.hstack((arr1,arr2)))                            #horizontal stack (axis = 1)
```
