```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,4*np.pi, 100)                         # we know that the period of sin is 2pi so that to cover the full sine wave we use 0 to 4pi 
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x,sin_y,label='Sine')
plt.plot(x,cos_y,label='Cosine')
plt.legend(loc = 'upper right')
plt.title('Sine Wave')
plt.show()
```

```python
plt.plot(x,sin_y,'*')
plt.plot(x,cos_y,"--")
plt.show()
```

```python
x = np.linspace(0,4*np.pi,100)
y = np.sin(x)
plt.style.use("Solarize_Light2")
plt.plot(x,y,label="Sine")
plt.legend()
plt.title("Sine Graph")
plt.show()
```

```python
x = np.linspace(0,4*np.pi,100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(15,5))
plt.style.use("Solarize_Light2")

plt.plot(x,y_sin,'*',color='red',linewidth=1,label='sin')
plt.plot(x,cos_y,'--',color='blue',linewidth=1,label='cos')

plt.legend()
plt.title("Sine and Cosine Graph")
plt.show()
```

```python
n = 500
x = np.random.randn(n)
y = np.random.randn(n)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.scatter(x,y,c='red',alpha=.5,s=30)                #alpha represents the transparency(visibility) and s represent the size 
plt.title("Scatter Plot")
plt.show()
```

```python
#Multiple Scatter Plot
n = 1000
x1 = np.random.randn(n)
y1 = np.random.randn(n)

x2 = np.random.randn(n)+2
y2 = np.random.randn(n)+2

plt.scatter(x1,y1,c='red',alpha=.5,s=20,label='Group 1')
plt.scatter(x2,y2,c='blue',alpha=.5,s=20,label="Group 2")
plt.legend()
plt.title("Scatter Plot with Multiple Groups")
plt.show()
```

The primary difference between plt.scatter and plt.plot is that plt.scatter gives us the freedom to set transparacy,size,color etc.

Pie Chart

```python
cars = ["TATA","LAND ROVER","MAHINDRA","HYNDUAI","MARUTI"]
data = [ 500  ,   50       ,    450   ,   550   ,   300  ]
plt.figure(figsize=(7,7))
plt.pie(data,labels = cars,wedgeprops={'edgecolor':'black'})
plt.title("Car Sales Distribution")
plt.show()
```

```python
cars = ["TATA","LAND ROVER","MAHINDRA","HYNDUAI","MARUTI"]
data = [ 500  ,   50       ,    450   ,   550   ,   300  ]
explode = [0,0,0,.2,.1]
plt.pie(data,labels=cars,wedgeprops={'edgecolor':'black','linewidth':2,'antialiased':True},explode=explode,shadow=True,autopct='%2.2f%%')
plt.show()
```

Multiple Subplots

```python
x = np.linspace(0,10,50)
fig,ax = plt.subplots(2,2)
y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sine Wave')

y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title("Cos Wave")

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title("Tan Wave")

y = np.random.randn(10000)
ax[1,1].hist(y,bins=40,edgecolor='black')
ax[1,1].set_title('Normal Distribution')
plt.tight_layout()
plt.show()
```
