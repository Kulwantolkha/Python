<font size = "6">Advanced Text Preprocessing</font>

<font size="2">If we want to change the pattern like we have "[7]" this type of pattern for allocation of refernce now if we want to remove these refernce from the whole text and want to print empty space in place of that then this could be done like this.</font>

```python
#References are from 7 to 51
for i in range(7,52):
    pattern = "[" + str(i) + "]"    #this is the pattern used here for reference
    txt = txt.replace(pattern," ")    
print(txt)
```

```python
#For special character
for i in "!@#$%^&*()":
    txt = txt.replace(i," ")
print(txt)
```

Writing in a file

```python
#Creating a file
fdw = open("Data1.txt",'w')
fdw.close()
```

```python
#Writing in a file
fdw = open("Data1.txt",'w')
fdw.write("Hello, How are you this text to be written in the Data1 file.")
fdw.close()
```

<font size="2">If we reopen the file in write mode only and write any data then the new data will be written on the earlier data it will erase the earlier data and write the new data. To add the data after that only we need to open the file in append mode.</font>

```python
fdw = open("Data1.txt",'a')
fdw.write("Hope the above text is successfully written in that file.")
fdw.close()
```

```python
#To write the data in new line
# - Open the file in append mode 
fdw = open("Data1.txt",'a')
fdw.write("\n Thanks!")
fdw.close()
```
