```python
#from above we can see that there are null values in rating,type,content rating, current ver, android ver
df.dropna(inplace=True,axis=0)
#This will delete all the null values row wise
```

```python
df.dropna(inplace=True,axis=1)
#This will delete all the null values column wise
```

```python
df.head()
print(len(df))
```

```python
link = 'https://raw.githubusercontent.com/AshishJangra27/Data-Analysis-with-Python-GFG/main/3.%20Data%20Preprocessing%20-%20Removing%20Null%20Value%20Rows/googleplaystore.csv'

df = pd.read_csv(link)

print(len(df))
df
```

```python
#now if we don't want to delete the null values we want to fill those values by some other values
#for integer values
missing_values = round(df['Rating'].mean(),2)            #this will fill the null values with the mean of average column
df['Rating'] = df['Rating'].fillna(missing_values)
df.isnull().sum()
```

```python
#for categorical values
df['Current Ver'] = df['Current Ver'].fillna('Varies on Device')
```
