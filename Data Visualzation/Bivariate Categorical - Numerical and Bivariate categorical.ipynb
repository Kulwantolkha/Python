```python
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset("titanic")
```

```python
df = data.copy()
```

```python
df.head()
```

```python
sns.barplot(data=df,x='class',y='age')
```

```python
sns.barplot(data=df,x='class',y='age',hue='sex')
```

```python
sns.barplot(data=df,x='class',y='age',hue='sex',estimator='sum')
```

```python
sns.boxplot(data=df,x='sex',y='age')
```

```python
sns.boxplot(data=df,x='sex',y='age',hue='class')
```

```python
sns.distplot(df[df['survived']==0]['age'],hist=False)
sns.distplot(df[df['survived']==1]['age'],hist=False)
```

```python
sns.violinplot(data=df,x='sex',y='fare')
```

<font size="12"> Bivariate Categorical</font>

```python
sns.barplot(data=df,x="sex",y="age",hue="survived")
```

```python
sns.stripplot(data=df,x="class",y="age",hue="sex")
```

```python
x=pd.crosstab(df["class"],df["survived"])
x
```

```python
sns.heatmap(x,annot=True,fmt=".1f")
```
